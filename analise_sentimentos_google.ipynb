```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
import pandas as pd
```

```python
df = pd.read_csv('articles_goog.csv')
```

```python
df
```

```python
# Carregar o FinBERT
finbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone", num_labels=3)
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")

# Criar o pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model=finbert, tokenizer=tokenizer)

# Notícias
data = df['title'].tolist()

# Analisar sentimentos
resultados = sentiment_pipeline(data)

# Adicionar as novas colunas ao DataFrame
df['sentimento'] = [resultado['label'] for resultado in resultados]
df['score'] = [resultado['score'] for resultado in resultados]

# Exibir o DataFrame atualizado
print(df)
```

```python
import pandas as pd

# Supondo que df já está definido e possui as colunas 'year_month', 'title', 'sentimento' e 'score'
# Exemplo de DataFrame
# df = pd.DataFrame({
#     'year_month': ['2023-02-01', '2023-02-02', '2023-03-01', '2023-03-02'],
#     'title': ['Notícia 1', 'Notícia 2', 'Notícia 3', 'Notícia 4'],
#     'sentimento': ['positivo', 'negativo', 'positivo', 'positivo'],
#     'score': [0.8, 0.6, 0.9, 0.7]
# })

# Converter a coluna 'year_month' para o formato datetime
df['year_month'] = pd.to_datetime(df['year_month'])

# Agrupar por mês e calcular a média dos scores
df_monthly_avg = df.groupby(df['year_month'].dt.to_period('M')).agg({
    'score': 'mean',
    'sentimento': lambda x: x.mode()[0]  # Sentimento mais frequente
}).reset_index()

# Renomear a coluna 'year_month' para 'month' para maior clareza
df_monthly_avg.rename(columns={'year_month': 'month'}, inplace=True)

# Exibir o DataFrame resultante
print(df_monthly_avg)
```
